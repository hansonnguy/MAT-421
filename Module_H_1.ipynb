# **22.1 ODE Initial Value Problem Statement**


---

Ordinary differential equations (ODEs) are a type of differential equation in which the independent variables, $x$, is one dimensional. We assume that an ODE can be written as $F(x,f(x),\frac{\mathrm{d}f(x)}{\mathrm{d}x},\frac{\mathrm{d}^2f(x)}{\mathrm{d}x^2},\frac{\mathrm{d}^3f(x)}{\mathrm{d}x^3},\dots,\frac{\mathrm{d}^{n-1}f(x)}{\mathrm{d}x^{n-1}})=\frac{\mathrm{d}^nf(x)}{\mathrm{d}x^n}$, where $F$ is some arbitrary function. The ODE is an $n^{th}$ order ODE. A general solution to a differential equation is a $g(x)$ that satisfies the differential equation. Generally stated, a particular solution $p(x)$ satisfies the differential equation and p^{(j)}(x_i)=p_i. A common set of known values for an ODE solution is the initial value. For an ODE of order $n$, we can have an initial value at $x=0,f(0),f(1)(0),f(2)(0),\dots,f(nâˆ’1)(0)$. If we have these values, we can set up our intial value problem.

# **22.2 Reduction of Order**


---

The ODE $F(x,f(x),\frac{\mathrm{d}f(x)}{\mathrm{d}x},\frac{\mathrm{d}^2f(x)}{\mathrm{d}x^2},\frac{\mathrm{d}^3f(x)}{\mathrm{d}x^3},\dots,\frac{\mathrm{d}^{n-1}f(x)}{\mathrm{d}x^{n-1}})=\frac{\mathrm{d}^nf(x)}{\mathrm{d}x^n}$ can be represented as $S(t) = \left[f(t), f^{(1)}(t), f^{(2)}(t), \dots, , f^{(n-1)}(t)\right]^T$. We can then take the derivative to obtain: $\frac{\mathrm{d}S(t)}{\mathrm{d}t} = \left[f^{(1)}(t), f^{(2)}(t), f^{(3)}(t), \dots, , f^{(n)}(t)\right]^T = \left[S_2(t), S_3(t), S_4(t), \dots, , F(t, S_1(t), S_2(t), \dots, S_{n-1}(t))\right]^T $, which is $n$ coupled first order ODEs. An ODE is considered linear if it can be written as: $\frac{\mathrm{d}S(t)}{\mathrm{d}t} = \left[ \begin{matrix}
a & b\\
c & d
\end{matrix} \right] S(t)$

# **22.3 The Euler Method**


---
The explicit Euler method approximates the solution of the intial value problem $S(t_{j+1})=S(t_j)+hF(t_j,S(t_j))$. The method is considered explicit because it is calculated directly with known values ($t_j$ and $S(t_j)$).  An example is given below:

```python
import numpy as np
import matplotlib.pyplot as plt


# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h_vals = [0.5, 0.1, 0.05, 0.01] # Step size
plt.figure(figsize = (12, 8))
for h in h_vals:
  t = np.arange(0, 1 + h, h) # Numerical grid
  s0 = -1 # Initial Condition

  # Explicit Euler Method
  s = np.zeros(len(t))
  s[0] = s0

  for i in range(0, len(t) - 1):
      s[i + 1] = s[i] + h*f(t[i], s[i])

  plt.plot(t, s, 'o--', label='h = ' + str(h))

plt.plot(t, -np.exp(-t), label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

The Implicit Euler Formula can be derived by taking the linear approximation of $S(t)$ around $t_{j+1}$ and computing it at $t_j$:

$S(t_{j+1})=S(t_j)+hF(t_{j+1},S(t_{j+1}))$.

However, this is difficult because $S(t_{j+1})$ appears on both sides of the equation. As a result, we must solve for $S(t_{j+1})$ through numerical iterative techniques.
